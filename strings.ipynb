```python
from math import gcd


class Solution:
    def leetcode1 (self,str1:str,str2:str):
        w1=len(str1)
        w2=len(str2)
        if (str1+str2 != str2+str1):
            return ""
        return str1[:gcd(w1,w2)]
```

```python
class SolutionAgain:
    def leetcode2(self,word1:str,word2:str):
        result=""
        w1=len(word1)
        w2=len(word2)
        for i in range(min(w1,w2)):
            result += word1[i]+word2[i]
        if w1> w2:
            result=result+word1[w2:]
        elif w2> w1:
            result=result+word2[w1:]
        return result
```

```python
class SolutionCandy:
    def candies(self,candies:list[int],extra:int):
        result=[]
        maxcandies=max(candies)
        for i in candies:
            result.append(i+extra >= maxcandies)
        return result
```

```python
sol3=SolutionCandy()
resultcandy=sol3.candies([1,2,3,4,5],3)
print(resultcandy)
```

```python
sol2=SolutionAgain()
result2=sol2.leetcode2("ABABC","KKK")
print(result2)
```
